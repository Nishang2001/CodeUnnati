```python
import numpy as np

x = np.array([2,3,8,7])
y = np.array([1,4,1,9])

plt.plot(x, y, marker = 'o', color = 'y', linestyle = 'dotted')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("chart")
plt.show()
```

```python
# import matplotlib.pyplot as plt
import random as ran

students = ["Jay","Nishang","Vinayak","Jyot","Raj"]

marks=[]
for i in range(0,len(students)):
    marks.append(ran.randint(0, 101))

plt.xlabel("Students")
plt.xticks(rotation=60)
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,marker = 'o', color = 'purple',markerfacecolor = 'y',markersize = 8)
```

```python
# importing the required module
import matplotlib.pyplot as plt
%matplotlib inline
# x axis values
x = [1,2,3]
# corresponding y axis values
y = [2,4,1]
# ploting the points
plt.plot(x, y)
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
# giving a title to my  graph
plt.title('My first graph!')
# function for show the plot
plt.show()
```

# Subplot

```python
# plot-1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)   # plt.subplot(rows, column, plot no.)
plt.plot(x,y)

# plot-2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2)
plt.plot(x,y)

# plot-3
x = np.array([0, 4, 2, 3])
y = np.array([10, 26, 37, 48])

plt.subplot(1, 3, 3)
plt.plot(x,y)

plt.suptitle("Subplotting")
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mec = 'r')
plt.show()
```

# Bar Plot

```python
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [2, 4, 6, 8, 10]

# y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.bar(x,y)

# function to show the plot
plt.show()
```

# Scatter Plot

```python
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 3, 6, 8, 1, 7, 10]

# y-axis values
y = [10, 5, 8, 4, 2, 1, 6, 3, 9, 7]

# Function to plot scatter
plt.scatter(x, y)

# function to show the plot
plt.title("Scatter Plot")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x-axis values
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

# y-axis values
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]

plt.plot(x, y, 'g')
plt.xlabel('x')
plt.ylabel('y')

plt.tick_params(axis="x", labelsize=18, labelrotation=-60, labelcolor="blue")
plt.tick_params(axis="y", labelsize=12, labelrotation=60, labelcolor="black")

plt.savefig("testchart",format='pdf')

plt.show()
```

```python
# Adding Legend in the graph
# importing modules
import numpy as np
import matplotlib.pyplot as plt

# Y-axis values
y1 = [2, 3, 4.5]

# Y-axis values
y2 = [1, 1.5, 5]

# Function to plot
plt.plot(y1,"blue")
plt.plot(y2,"red")

# function add a legend
plt.legend(["blue", "red"], loc = "upper left")
plt.show()
```

# Histogram

```python
data = np.random.normal(170, 10, 250)
#print(data)

# The hist() function will read the array and produce a histogram:
# set the bin value 30
plt.hist(data);

plt.show()
plt.hist(data, bins=30);

plt.show()
```

```python
fix, ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

# text annotation to the plot where it indicate maximum value of the curve
ax.annotate('local maximum', xy=(6.28,1), xytext=(10,4), arrowprops=dict(facecolor='black', shrink=0.05))

# text annotation to the plot where it indicate maximum value of the curve
ax.annotate('local minimum', xy=(5*np.pi,-1), xytext=(2,-6), arrowprops=dict(arrowstyle='->', connectionstyle="angle3,angleA=0,angle=-90"))
```
